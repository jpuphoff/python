## Pandas 

```python
import pandas as pd
```

The two most important types of data structure in PANDAS are:
* Series: one-dimensional labeled array capable of holding data of any type, labeled with an index (it's basically a column)

```python
print(pd.Series([1,1,5,"hi",2,"ohlala"]))
```

```python
s = pd.Series({"Ana":2,"Maria":5,"Carmen":3})     
print(s["Ana"])
print(s[1])
```

* Dataframe: 2-dimensional labeled data structure with columns of potentially different types

```python
print(pd.DataFrame([[3,2,5],[9,8,7]]))
```

```python
df = {"ana":[1991,16,10],"sophie":[2008,14,6],"carmen":[1967,5,5]}
print(pd.DataFrame(df,index = ["Year","Day","Month"]))
```

## Matplotlib

```python
import matplotlib 
import numpy as np
import matplotlib.pyplot as plt
```

#### Graph 1

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.1)

plt.plot(x, np.cos(x), "r", x,np.sin(x),"g")       #plot & colors
plt.xlabel('x')                                    #labels
plt.ylabel('cos(x)')
plt.show()                                         #show plot
```

#### Graph 2

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('Whatever')
plt.ylabel('Y')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel("X")
plt.ylabel('Y')

plt.show()
```

#### Graph 3

```python
np.random.seed(19680801)
data = np.random.randn(2, 100)

fig, axs = plt.subplots(2, 2, figsize=(5, 5))
axs[0, 0].hist(data[0])
axs[1, 0].scatter(data[0], data[1])
axs[0, 1].plot(data[0], data[1])
axs[1, 1].hist2d(data[0], data[1])

plt.show()
```

#### Graph 4: Histogram

```python
mu, sigma = 50, 7
x = mu + sigma * np.random.randn(10000)
n, bins, patches = plt.hist(x, 20, density=1, facecolor='dodgerblue', alpha=0.6)  

plt.xlabel('x')
plt.ylabel('Probability')
plt.text(20, .045, r'$\mu=50,\ \sigma=7$')  #show value of mu and sigma (first values for position)
plt.axis([10, 80, 0, 0.07])                 #size of axis:xmin=10,xmax=80 and ymin=0, ymax=0.07
plt.grid(True)
plt.show()
```

#### Graph 5: Pie Chart


```python
labels = 'Chocolate', 'Wheat', 'Sugar', 'Egg'
sizes = [15, 30, 45, 10]
explode = (0, 0, 0.05, 0)        #"explode" the 3rd slice

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=False, startangle=50)
plt.show()
```
